```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file = pd.read_csv('salary.csv')
file
```

```python
null = file[file['salary'].isnull()]
print(null)

print('The missing value is in column Salary in index 208')

```

```python
exprior = file['exprior']. sum()
print("There are ", str(exprior), "people that were surveyed.")
```

```python
# here we are extracting the salary column so that we can focus on it
salary  = file['salary']


# here we will find the mean salary using the numpy fynction mean 
mean = np.mean(salary)
print("The Mean salary as an integer, without decimal points is", str(int(mean)), '\n')

# in order for us to get the lowest and highest salary we have to sort the salary dataframe (column).
salary = sorted(salary)

# print the lowest and highest salary as integers (without decimal points)

print("The lowest salary as an integer, without decimal points is", str(int(salary[0])))
print("The lowest salary as an integer, without decimal points is", str(int(salary[-1])))
print('\n')

# print the lowest and highest salary in its original form

print("The lowest salary is", str(salary[0]))
print("The lowest salary is", str(salary[-1]))
print('\n')
```

```python
# We will extract the Years Worked column so that we can get the standard  deviation for the years worked

yearsworked = file['yearsworked']

# getting the standard  deviation using numpy 

std = np.std(yearsworked)

# Printing it out as an integer (without decimal points)

print("The Standard deviation of the years worked as an integer, without decimal points is ", str(int(std)))
print('\n')

# Printing it out std in its original form

print("The Standard deviation of the years worked is", str(std))
```

```python
import scipy.stats
scipy.stats.iqr(salary)
```

```python
gender = file['male']
males = gender[gender == 1] .count()

print("The number of males in the sample is ", str(males))
print('\n')

females = gender[gender == 0] .count()

print("The number of males in the sample is ", str(females))

df  = males, females

labels = ['Males', 'Females', ]
color = ['palegreen','crimson']

#show percentage for each group on the pie chart 

plt.pie(df, labels =labels, colors = color, autopct='%1.0f%%')
plt.title('Number of females and males in the sample')
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.show()
```

```python
femaledegree = file[file.male == 0]

degree = femaledegree['degree'] .count()
print('A ', str(degree), "women have Degrees")
```

```python
salary = file.groupby('salary') .size() .reset_index(name = 'frq')

plt.scatter( salary['frq'],salary['salary'])
plt.ylabel('Salaries')
plt.xlabel('Number of responses')
plt.title('Salaries of people')
plt.show()
```
